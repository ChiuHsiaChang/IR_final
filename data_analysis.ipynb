```python
import pandas as pd
import numpy as np

df = pd.read_csv('./train_ori.csv')
```

```python
print(np.mean(df.q.apply(lambda x: len(x))))
print(np.mean(df.r.apply(lambda x: len(x))))
```

```python
def len_word(sentence):
    s_len = len(sentence.split())
    return s_len
```

```python
print(len_word(df.r[0]))
```

```python
total_num = df.shape[0]
q_512 = df.q.apply(lambda x: len_word(x) > 512).value_counts()
r_512 = df.r.apply(lambda x: len_word(x) > 512).value_counts()
q_ratio = q_512/total_num
r_ratio = r_512/total_num
print(q_512,q_ratio,'\n')
print(r_512,r_ratio)
```

```python
max_len_q = 0
for i in range(0,total_num):
    len_qi = len_word(df.q[i])
    if len_qi>max_len_q:
        max_len_q = len_qi
print(max_len_q)
```
